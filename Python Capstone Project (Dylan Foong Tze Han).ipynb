# Data Cleaning
## Dataset 1 - Facility Data

```python
# Create facility_df_summary
facility_df_summary = facility_df.groupby('FINANCIAL_APPLICATION_RK').agg(
    APPLIED_AMT_TOTAL=('APPLIED_AMT', 'sum'),
    APPROVED_LOAN_AMT_TOTAL=('APPROVED_LOAN_AMT', 'sum')
).reset_index()

# Left join facility_df_summary back into the original facility_df
facility_df = facility_df.merge(facility_df_summary, on = 'FINANCIAL_APPLICATION_RK', how = 'left')

# Keep the first record for each FINANCIAL_APPLICATION_RK after sorting
facility_df = facility_df.sort_values(by=['FINANCIAL_APPLICATION_RK', 'FACILITY_ID']).drop_duplicates(subset = 'FINANCIAL_APPLICATION_RK', keep = 'first')

print("----------------------------Final Facility Data----------------------------")
display(facility_df)

facility_df.to_csv('facility_data_python.csv', index = False)
```

## Dataset 2 - Applicant Data

```python
# To check data types and ensure data types as per requirement
print(applicant_df.dtypes)

# Calculate the age
applicant_df['AGE'] = applicant_df['X_UPDATE_DTTM'].dt.year - applicant_df['BIRTH_DT'].dt.year

# To ensure the specified columns are float
float_columns = [
    'X_BASIC_INCOME_AMT',
    'X_FIXED_ALLOWANCE_AMT',
    'X_TOTAL_VARIABLE_INCOME_AMT',
    'X_VARIABLE_OVERTIME_AMT',
    'X_TAX_AMT',
    'X_STATUTORY_DEDUCTIONS_EPF',
    'X_STATUTORY_DEDUCTIONS_SOCSO'
]

# Convert the specified columns to float
applicant_df[float_columns] = applicant_df[float_columns].astype(float)

# Replace missing values with 0 in the specified columns
applicant_df[float_columns] = applicant_df[float_columns].fillna(0)
applicant_df['NET_INCOME'] = (
    applicant_df["X_BASIC_INCOME_AMT"] + 
    applicant_df["X_FIXED_ALLOWANCE_AMT"] + 
    applicant_df["X_TOTAL_VARIABLE_INCOME_AMT"] + 
    applicant_df["X_VARIABLE_OVERTIME_AMT"] - 
    applicant_df["X_TAX_AMT"] - 
    applicant_df["X_STATUTORY_DEDUCTIONS_EPF"] - 
    applicant_df["X_STATUTORY_DEDUCTIONS_SOCSO"]
)

# Derive EMPLOYMENT_YEARS_CNT
def calculate_employment_years(row):
    if row['X_GAP_CURR_PREV_EMPLOY_MTH_CNT'] <= 12:
        return row['CURR_EMPLOYMENT_YEARS_CNT'] + row['X_PREV_EMPLOY_YEARS_CNT']
    else:
        return row['CURR_EMPLOYMENT_YEARS_CNT']

applicant_df['EMPLOYMENT_YEARS_CNT'] = applicant_df.apply(calculate_employment_years, axis = 1)

# Check if X_ASCORE_SEQ in applicant data is also in bureau data
applicant_df['linkage'] = applicant_df['X_ASCORE_SEQ'].isin(bureau_raw['X_ASCORE_SEQ'])

# Derive CCRIS_IND
def result(row):
    if pd.notnull(row['X_ASCORE_SEQ']):
        if row['linkage']:
            return 'With CCRIS Commitment'
        else:
            return 'Linkage Issues'
    else:
        return 'Without CCRIS Commitment'

applicant_df['CCRIS_IND'] = applicant_df.apply(result, axis=1)

# Sort X_UPDATE_DTTM in decending order and set FINANCIAL_APPLICATION_RK as index to remove duplicate APPLICANT_RK
applicant_df = applicant_df.sort_values(by = ['X_UPDATE_DTTM'], ascending = False).drop_duplicates(subset = 'APPLICANT_RK', keep = 'first').drop(columns = ['linkage'])

print("----------------------------Final Applicant Data----------------------------")
display(applicant_df)
applicant_df.to_csv('applicant_data_python.csv', index = False)
```

## Dataset 3 - Bureau Data

```python
# Convert the specified columns to string
bureau_df['LEGAL_STATUS_CD'] = bureau_df['LEGAL_STATUS_CD'].astype(str)

# To check data types and ensure data types as per requirement
display(bureau_df.dtypes)

# Derive FACILITY_NO
bureau_df['FACILITY_NO'] = bureau_df.groupby(['X_ASCORE_SEQ', 'SECTION_KEY', 'ROW_NO']).cumcount() + 1

# Derive CC_LE1000
def credit_card_less1000(row):
    if row['SECTION_KEY'] == "Outstanding Credit":
        if row['FACILITY_CD'] == "CRDTCARD" and row['BALANCE_AMT'] <= 1000:
            return "Y"

bureau_df['CC_LE1000'] = bureau_df.apply(credit_card_less1000, axis = 1)

# Derive WEIGHTED_AVG_LIMIT_AMT
bureau_df['SUM_BALANCE'] = bureau_df.groupby(['X_ASCORE_SEQ', 'SECTION_KEY', 'ROW_NO'])['BALANCE_AMT'].transform('sum')
# nunique computes how many unique values within each group defined
bureau_df['TOTAL_FACILITY_NO'] = bureau_df.groupby(['X_ASCORE_SEQ', 'SECTION_KEY', 'ROW_NO'])['FACILITY_NO'].transform('nunique')
# assign new as a column, then group by new to do
bureau_df['TOTAL_CC_NO'] = bureau_df.assign(new = bureau_df['FACILITY_NO'].where(bureau_df.FACILITY_CD.eq('CRDTCARD'))).groupby(['X_ASCORE_SEQ', 'SECTION_KEY', 'ROW_NO'])['new'].transform('nunique')
bureau_df['TOTAL_BALANCE_0'] = bureau_df.assign(new = bureau_df['FACILITY_NO'].where(bureau_df.BALANCE_AMT.eq(0))).groupby(['X_ASCORE_SEQ', 'SECTION_KEY', 'ROW_NO'])['new'].transform('nunique')

def compute_weighted_avg_limit_amt(row):
    if row['SECTION_KEY'] == "Outstanding Credit":
        if row['TOTAL_FACILITY_NO'] == 1:
            # Only one FACILITY_NO within ROW_NO
            return row['LIMIT_AMT']
        elif row['TOTAL_FACILITY_NO'] == row['TOTAL_BALANCE_0']:
            # All FACILITY_NO within ROW_NO have BALANCE_AMT = 0
            return row['LIMIT_AMT'] / row['TOTAL_FACILITY_NO']
        elif row['TOTAL_FACILITY_NO'] == row['TOTAL_CC_NO'] and row['SUM_BALANCE'] < row['LIMIT_AMT']:
            # All FACILITY_NO within ROW_NO have FACILITY_CD = CRDTCARD
            return row['BALANCE_AMT'] / row['SUM_BALANCE'] * row['LIMIT_AMT']
    else:
        return np.nan

# Apply the function
bureau_df['WEIGHTED_AVG_LIMIT_AMT'] = bureau_df.apply(compute_weighted_avg_limit_amt, axis = 1)
bureau_df = bureau_df.drop(['SUM_BALANCE', 'TOTAL_FACILITY_NO', 'TOTAL_CC_NO', 'TOTAL_BALANCE_0'], axis = 1)

# Derive APP_MONTH_ON_BOOK
def diff_in_months(row):
    return (row['Bureau_Report_Pull_Dt'].year - row['APPROVAL_DT'].year) * 12 + row['Bureau_Report_Pull_Dt'].month - row['APPROVAL_DT'].month

bureau_df['APP_MONTH_ON_BOOK'] = bureau_df.apply(diff_in_months, axis = 1) + 1

# List to store the column names
months = [f'ACCOUNT_CONDUCT_M{i}' for i in range(1, 13)]

# Create a new column MAX_DELINQ with maximum value from columns ACCOUNT_CONDUCT_M1 to ACCOUNT_CONDUCT_M12
def max_delinquency(row):
    global months
    if row['SECTION_KEY'] == "Outstanding Credit":
        return row[months].max()

bureau_df['MAX_DELINQ'] = bureau_df.apply(max_delinquency, axis = 1)
# bureau_df['MAX_DELINQ'] = filtered_bureau[months].max(axis = 1)

# Derive DOUBLE2
def count_consecutive_doubles(row):
    global months
    if row[months].isna().all():
        return np.nan
    double2 = 0
    i = 0
    while i < len(months) - 1:
        if row[months[i]] >= 2 and row[months[i+1]] >= 2:
            double2 += 1
            i += 2
        else:
            i += 1
    return double2

bureau_df['DOUBLE2'] = bureau_df.apply(count_consecutive_doubles, axis = 1)

# Derive TRIPLE2
def count_once_triple2(row):
    global months
    if row[months].isna().all():
        return np.nan
    triple2 = 0
    i = 0
    while i < len(months) - 1:
        if row[months[i]] >= 2 and row[months[i+1]] >= 2 and row[months[i+2]] >= 2:
            triple2 = 1
            i += 3
        else:
            i += 1
    return triple2

bureau_df['TRIPLE2'] = bureau_df.apply(count_once_triple2, axis = 1)

# Derive MTH_SNC_MAX_DLQ
def calculate_mth_snc_max_dlq(row):
    conduct_values = row[months].values
    non_null_indices = [i for i, val in enumerate(conduct_values) if pd.notnull(val)]
    if not non_null_indices:
        return np.nan
    first_non_null_month = non_null_indices[0]
    
    if row["MAX_DELINQ"] == 0:
        return 12
    
    max_delinquency_month = next(i for i, val in enumerate(conduct_values) if val == row["MAX_DELINQ"])
    return max_delinquency_month - first_non_null_month

bureau_df['MTH_SNC_MAX_DLQ'] = bureau_df.apply(calculate_mth_snc_max_dlq, axis = 1)

# Derive SUM_DOUBLE2
bureau_df_agg = bureau_df.groupby('X_ASCORE_SEQ').agg(SUM_DOUBLE2=('DOUBLE2', 'sum')).reset_index()
bureau_df_agg = bureau_df.merge(bureau_df_agg, on = 'X_ASCORE_SEQ', how = 'left')

print("----------------------------Final Bureau Data----------------------------")
display(bureau_df_agg)
bureau_df_agg.to_csv('bureau_data_python.csv', index = False)
```

## Dataset 4 - Performance Data

```python
performance_df['OPEN_DT'] = pd.to_datetime(performance_df['OPEN_DT'])

# To check data types and ensure data types as per requirement
display(performance_df.dtypes)

# Derive GB_IND
def good_bad_indicator(row):
    if row['ACCOUNT_STATUS_CD'] == "W":
        return "Written-Off"
    if row['ACCOUNT_STATUS_CD'] in ("C", "T"):
        return "Restructuring & Rescheduling (R&R)"
    if row['ACCOUNT_STATUS_CD'] == "K":
        return "AKPK"
    if row['X_IMPAIRED_FLG'] == "Y":
        return "Impaired Loans"    
    if row['PAYMENT_TIME_FREQUENCY_CD'] in (1, 6) and row['X_ECCRIS_DIA'] >= 90:
        return "DIA >= 90"
    else:
        return "N"

performance_df['GB_IND'] = performance_df.apply(good_bad_indicator, axis = 1)

# Bring over PERF_START_DT and PERF_WINDOW from Facility Data
new_facility_df = facility_df[['FINANCIAL_APPLICATION_RK', 'PERF_START_DT', 'PERF_WINDOW']]
performance_df = performance_df.merge(new_facility_df, on = 'FINANCIAL_APPLICATION_RK', how = 'left')

# Derive PERF_END_DT
performance_df['PERF_END_DT'] = performance_df.apply(lambda row: row['PERF_START_DT'] + pd.DateOffset(months=row['PERF_WINDOW']), axis=1)

# Filter rows where d_pou is not within the PERF_START_DT and PERF_END_DT
performance_df = performance_df[(performance_df['d_pou'] >= performance_df['PERF_START_DT']) & (performance_df['d_pou'] <= performance_df['PERF_END_DT'])]

# The given dictionary for mapping GB_IND to ranks
gb_ind_dict = {
    'Written-Off': 1,
    'Restructuring & Rescheduling (R&R)': 2,
    'AKPK': 3,
    'Impaired Loans': 4,
    'DIA >= 90': 5,
    'N': 6
}

# Derive GB_IND_RANK column using the dictionary
performance_df['GB_IND_RANK'] = performance_df['GB_IND'].map(gb_ind_dict)
display(performance_df)

# Aggregate to ACCOUNT_RK level by keeping the row with the lowest GB_IND_RANK for each ACCOUNT_RK
performance_df_agg = performance_df.loc[performance_df.groupby('ACCOUNT_RK')['GB_IND_RANK'].idxmin()]

print("----------------------------Final Performance Data----------------------------")
display(performance_df_agg)
performance_df_agg.to_csv('performance_data_python.csv', index = False)
```
